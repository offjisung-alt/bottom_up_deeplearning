# 1. 완전 연결 계층으로 이루어진 네트워크

일반적인 신경망은 완전 연결 계층, 즉 Affine 계층을 중심으로 구성된다.

구조는 다음과 같다.

```text
Input
↓
Affine
↓
ReLU
↓
Affine
↓
ReLU
↓
...
↓
Affine
↓
Softmax
```

---

## Affine 계층이란?

Affine 계층은 입력 데이터에 가중치를 곱하고 편향을 더하는 계층이다.

$$
Y = XW + b
$$

여기서

- $X$ : 입력 데이터
- $W$ : 가중치
- $b$ : 편향
- $Y$ : 출력값

이다.

---

## 예시

이미지 데이터가 있다고 하자.

예를 들어 흑백 이미지 한 장의 크기가

$$
28 \times 28
$$

이면 픽셀 수는

$$
28 \times 28 = 784
$$

이다.

완전 연결 계층에서는 이 이미지를 다음과 같이 1차원 벡터로 펼친다.

$$
28 \times 28
\rightarrow
784
$$

즉, 이미지의 2차원 구조가 사라진다.

---

# 2. 완전 연결 계층의 문제점

완전 연결 계층의 가장 큰 문제는

```text
데이터의 형상이 무시된다
```

는 것이다.

---

이미지는 원래 다음과 같은 구조를 가진다.

```text
높이 Height
너비 Width
채널 Channel
```

예를 들어 컬러 이미지는 보통

$$
3 \times 32 \times 32
$$

형태를 가진다.

여기서

- $3$ : 채널 수, RGB
- $32$ : 높이
- $32$ : 너비

이다.

---

하지만 Affine 계층에 넣기 위해 이미지를 1차원으로 펼치면

$$
3 \times 32 \times 32
=
3072
$$

개의 숫자로 바뀐다.

이 과정에서 다음 정보가 약해진다.

- 가까운 픽셀끼리의 관계
- 가로, 세로 방향의 패턴
- 채널 사이의 관계
- 이미지의 공간적 구조

---

# 3. CNN으로 이루어진 네트워크

CNN은 Convolutional Neural Network의 약자이다.

이미지처럼 공간 구조가 중요한 데이터를 처리하기 위해 만들어진 신경망이다.

CNN의 기본 구조는 다음과 같다.

```text
Input
↓
Conv
↓
ReLU
↓
Pooling
↓
Conv
↓
ReLU
↓
Pooling
↓
...
↓
Affine
↓
ReLU
↓
Affine
↓
Softmax
```

---

## CNN의 핵심

CNN은 이미지를 1차원으로 바로 펼치지 않는다.

대신 이미지의 높이, 너비, 채널 구조를 유지한 채로 학습한다.

즉,

```text
이미지의 공간 구조를 보존하면서 특징을 추출한다
```

는 것이 CNN의 핵심이다.

---

# 4. 합성곱 계층 Convolution Layer

합성곱 계층은 CNN에서 가장 중요한 계층이다.

합성곱 계층은 작은 필터를 이미지 위에서 이동시키면서 특징을 추출한다.

---

## 필터 또는 커널

합성곱에서 사용하는 작은 가중치 행렬을

```text
필터 Filter
또는
커널 Kernel
```

이라고 한다.

예를 들어 $3 \times 3$ 필터는 다음과 같다.

$$
\begin{bmatrix}
w_{11} & w_{12} & w_{13}\\
w_{21} & w_{22} & w_{23}\\
w_{31} & w_{32} & w_{33}
\end{bmatrix}
$$

---

## 윈도우 Window

필터가 입력 데이터 위에서 한 번에 바라보는 영역을 윈도우라고 한다.

예를 들어 $3 \times 3$ 필터를 사용하면 입력 이미지의 $3 \times 3$ 영역을 한 번에 본다.

```text
입력 이미지의 일부 영역
↓
필터와 곱셈
↓
합산
↓
출력값 하나 생성
```

---

# 5. 합성곱 연산

입력 데이터의 일부 영역과 필터를 원소별로 곱한 뒤 모두 더한다.

예를 들어 입력의 한 영역이

$$
\begin{bmatrix}
1 & 2 & 3\\
4 & 5 & 6\\
7 & 8 & 9
\end{bmatrix}
$$

이고 필터가

$$
\begin{bmatrix}
1 & 0 & 1\\
0 & 1 & 0\\
1 & 0 & 1
\end{bmatrix}
$$

이면 합성곱 결과는 다음과 같다.

$$
1 \times 1
+
2 \times 0
+
3 \times 1
+
4 \times 0
+
5 \times 1
+
6 \times 0
+
7 \times 1
+
8 \times 0
+
9 \times 1
$$

$$
=
1+3+5+7+9
=
25
$$

이 값 하나가 출력 특징 맵의 한 칸이 된다.

---

# 6. 특징 맵 Feature Map

필터를 이미지 전체에 이동시키면서 합성곱 연산을 수행하면 출력 데이터가 만들어진다.

이 출력 데이터를 특징 맵이라고 한다.

```text
입력 이미지
↓
필터 적용
↓
특징 맵 생성
```

---

## 여러 개의 필터

필터가 여러 개 있으면 특징 맵도 여러 개 만들어진다.

예를 들어 필터가 16개라면 출력 특징 맵도 16개이다.

즉, 출력 채널 수는 필터 개수와 같다.

---

# 7. 패딩 Padding

패딩은 입력 데이터의 가장자리에 값을 추가하는 것이다.

보통 0을 추가하므로 Zero Padding이라고 한다.

예를 들어

$$
4 \times 4
$$

입력에 패딩 1을 적용하면

$$
6 \times 6
$$

이 된다.

---

## 패딩을 사용하는 이유

### 1. 출력 크기 조절

패딩이 없으면 합성곱을 할 때마다 출력 크기가 작아진다.

패딩을 사용하면 출력 크기를 유지할 수 있다.

---

### 2. 가장자리 정보 보존

패딩이 없으면 이미지의 가장자리 픽셀은 합성곱에 적게 참여한다.

패딩을 추가하면 가장자리 정보도 충분히 반영할 수 있다.

---

# 8. 스트라이드 Stride

스트라이드는 필터가 이동하는 간격이다.

예를 들어

```text
stride = 1
```

이면 필터가 한 칸씩 이동한다.

```text
stride = 2
```

이면 필터가 두 칸씩 이동한다.

---

## 스트라이드가 커지면?

스트라이드가 커질수록 출력 크기는 작아진다.

```text
stride 작음 → 출력 크기 큼
stride 큼 → 출력 크기 작음
```

---

# 9. 출력 크기 공식

입력 크기를

$$
H \times W
$$

필터 크기를

$$
FH \times FW
$$

패딩을

$$
P
$$

스트라이드를

$$
S
$$

라고 하자.

출력 높이와 출력 너비는 다음과 같다.

$$
OH
=
\frac{H+2P-FH}{S}
+
1
$$

$$
OW
=
\frac{W+2P-FW}{S}
+
1
$$

여기서

- $H$ : 입력 높이
- $W$ : 입력 너비
- $FH$ : 필터 높이
- $FW$ : 필터 너비
- $P$ : 패딩 크기
- $S$ : 스트라이드
- $OH$ : 출력 높이
- $OW$ : 출력 너비

이다.

---

## 출력 크기 예시

입력 크기

$$
H=7,\quad W=7
$$

필터 크기

$$
FH=3,\quad FW=3
$$

패딩

$$
P=0
$$

스트라이드

$$
S=1
$$

이면

$$
OH=
\frac{7+2 \times 0-3}{1}+1
=
5
$$

$$
OW=
\frac{7+2 \times 0-3}{1}+1
=
5
$$

따라서 출력 크기는

$$
5 \times 5
$$

이다.

# 10. 3차원 데이터의 합성곱 연산

이미지는 보통 3차원 데이터이다.

예를 들어 컬러 이미지는 다음과 같다.

$$
C \times H \times W
$$

여기서

- $C$ : 채널 수
- $H$ : 높이
- $W$ : 너비

이다.

---

## RGB 이미지 예시

컬러 이미지의 경우 채널이 3개이다.

```text
Red 채널
Green 채널
Blue 채널
```

따라서 입력 데이터는 다음과 같은 형상이다.

$$
3 \times H \times W
$$

---

## 3차원 필터

3차원 입력 데이터에 적용하는 필터도 채널 수가 같아야 한다.

입력이

$$
3 \times 32 \times 32
$$

이면 필터도 채널 방향으로 3개의 깊이를 가진다.

예를 들어 필터 크기가 $3 \times 3$이면 실제 필터 형상은

$$
3 \times 3 \times 3
$$

이다.

보통 다음과 같이 표현한다.

$$
C \times FH \times FW
$$

---

## 3차원 합성곱 과정

각 채널별로 합성곱을 수행한 뒤 결과를 모두 더한다.

```text
Red 채널 합성곱
+
Green 채널 합성곱
+
Blue 채널 합성곱
+
Bias
=
출력값 하나
```

---

수식으로 표현하면 다음과 같다.

$$
y
=
\sum_{c=1}^{C}
\sum_{i=1}^{FH}
\sum_{j=1}^{FW}
x_{c,i,j}w_{c,i,j}
+
b
$$

---

## 출력 채널 수

출력 채널 수는 필터 개수에 의해 결정된다.

예를 들어 입력이

$$
3 \times 32 \times 32
$$

이고 필터가 16개라면 출력은

$$
16 \times OH \times OW
$$

가 된다.

즉,

```text
필터 1개 → 특징 맵 1개
필터 16개 → 특징 맵 16개
필터 32개 → 특징 맵 32개
```

이다.

---

# 11. 배치 처리

딥러닝에서는 데이터를 한 장씩 처리하지 않고 여러 장을 묶어서 처리한다.

이 묶음을 배치 Batch라고 한다.

---

## 배치 데이터 형상

CNN에서 입력 데이터는 보통 다음 형태를 가진다.

$$
N \times C \times H \times W
$$

여기서

- $N$ : 배치 크기
- $C$ : 채널 수
- $H$ : 높이
- $W$ : 너비

이다.

---

## 예시

이미지 32장을 한 번에 처리하고, 각 이미지가 RGB 이미지이며 크기가 $64 \times 64$라면 입력 형상은

$$
32 \times 3 \times 64 \times 64
$$

이다.

---

## 필터 형상

필터는 보통 다음과 같은 형상을 가진다.

$$
FN \times C \times FH \times FW
$$

여기서

- $FN$ : 필터 개수
- $C$ : 입력 채널 수
- $FH$ : 필터 높이
- $FW$ : 필터 너비

이다.

---

## 출력 형상

출력 형상은 다음과 같다.

$$
N \times FN \times OH \times OW
$$

예를 들어

입력

$$
32 \times 3 \times 64 \times 64
$$

필터

$$
16 \times 3 \times 3 \times 3
$$

패딩 1, 스트라이드 1이면 출력은

$$
32 \times 16 \times 64 \times 64
$$

이다.

---

# 12. 풀링 계층 Pooling Layer

풀링 계층은 공간 크기를 줄이는 계층이다.

주로 합성곱 계층 뒤에 사용된다.

```text
Conv
↓
ReLU
↓
Pooling
```

---

## 풀링의 목적

### 1. 데이터 크기 감소

특징 맵의 높이와 너비를 줄인다.

예를 들어

$$
4 \times 4
$$

특징 맵에 $2 \times 2$ 풀링을 적용하면

$$
2 \times 2
$$

가 된다.

---

### 2. 중요한 특징 유지

작은 영역에서 대표값을 뽑아 중요한 특징을 유지한다.

---

### 3. 위치 변화에 둔감

물체가 조금 이동해도 비슷한 특징을 얻을 수 있다.

---

# 13. Max Pooling

가장 많이 사용하는 풀링 방법이다.

Max Pooling은 작은 영역에서 가장 큰 값을 선택한다.

---

## 예시

입력 영역이 다음과 같다고 하자.

$$
\begin{bmatrix}
1 & 3\\
2 & 4
\end{bmatrix}
$$

이면 Max Pooling 결과는

$$
4
$$

이다.

---

## $2 \times 2$ Max Pooling 예시

입력 특징 맵이

$$
\begin{bmatrix}
1 & 3 & 2 & 1\\
2 & 4 & 1 & 0\\
0 & 1 & 5 & 2\\
1 & 2 & 3 & 4
\end{bmatrix}
$$

이고 $2 \times 2$ Max Pooling을 적용하면

$$
\begin{bmatrix}
4 & 2\\
2 & 5
\end{bmatrix}
$$

이 된다.

---

# 14. Average Pooling

Average Pooling은 작은 영역의 평균값을 선택한다.

예를 들어

$$
\begin{bmatrix}
1 & 3\\
2 & 4
\end{bmatrix}
$$

이면

$$
\frac{1+3+2+4}{4}=2.5
$$

가 된다.

---

# 15. 풀링 계층의 특징

## 1. 학습해야 할 매개변수가 없다

풀링은 최댓값이나 평균값을 계산할 뿐이다.

따라서 가중치가 없다.

---

## 2. 채널 수가 변하지 않는다

풀링은 각 채널별로 독립적으로 적용된다.

입력이

$$
16 \times 32 \times 32
$$

이면 $2 \times 2$ 풀링 후 출력은

$$
16 \times 16 \times 16
$$

이다.

채널 수 16은 그대로 유지된다.

---

## 3. 위치 변화에 강하다

이미지 안의 물체가 조금 이동해도 풀링 결과는 크게 변하지 않는다.

---

# im2col 함수

im2col은 CNN의 합성곱(Convolution) 연산을 빠르게 수행하기 위해 사용하는 기법이다.

im2col은

```text
image to column
```

의 약자이다.

즉,

```text
이미지 데이터를
행렬(Column 형태)로 변환하는 함수
```

이다.

---

# 1. 왜 im2col이 필요한가?

CNN의 합성곱 연산은 계산량이 매우 많다.

예를 들어

입력 이미지

$$
28 \times 28
$$

필터

$$
5 \times 5
$$

가 있다고 하자.

---

합성곱 연산은

```text
필터를 한 칸 이동
↓
곱셈
↓
덧셈

또 이동
↓
곱셈
↓
덧셈
```

을 반복한다.

즉,

```python
for y:
    for x:
        for filter:
            ...
```

처럼 반복문이 매우 많이 사용된다.

---

문제는

```text
반복문은 느리다
```

는 것이다.

특히 Python에서는 더욱 느리다.

---

그래서

```text
합성곱
↓
행렬 곱셈
```

으로 바꾸어 계산한다.

이를 위해 사용하는 기법이

```text
im2col
```

이다.

---

# 2. im2col의 핵심 아이디어

합성곱 연산을

$$
\text{Convolution}
$$

에서

$$
\text{Matrix Multiplication}
$$

으로 변환한다.

---

왜 행렬 곱셈으로 바꾸는가?

NumPy와 GPU는 행렬 곱셈을 매우 빠르게 수행할 수 있기 때문이다.

즉

```text
느린 반복문 제거
↓
빠른 행렬 곱 사용
↓
CNN 속도 향상
```

이 된다.

# 9. im2col 결과 형상

필터 크기를

$$

$$

$$

\frac{W+2P-FW}{S}+1
$$

---

그러면

im2col 결과 형상은

$$
(N \times OH \times OW,
C \times FH \times FW)
$$

가 된다.

# 핵심 정리

1. im2col은 Image to Column의 약자이다.

2. 합성곱 연산을 행렬 곱으로 변환하는 기법이다.

3. 모든 윈도우를 펼쳐 하나의 큰 행렬로 만든다.

4. 필터도 펼쳐 행렬 형태로 만든다.

5. 이후 행렬 곱으로 합성곱을 수행한다.

6. 일반적인 im2col 결과 형상은

$$
(N\times OH\times OW,
C\times FH\times FW)
$$

이다.

7. 역변환은 col2im이 수행한다.

8. 속도는 매우 빨라지지만 메모리를 더 많이 사용한다.

````python
x1 = np.random.rand(1,3,7,7) # (데이터 수, 채널 수, 높이)
col1 = im2col(x1, 5, 5, stride =1, pad =0)
print(col1.shape)

x2 = np.random.rand(10,3,7,7) # (데이터 수, 채널 수, 높이)
col2 = im2col(x2, 5, 5, stride =1, pad =0)
print(col2.shape)

#shape -> (데이터 수 x 출력의 높이 x 출력의 너비 , 채널 수 x 필터의 높이 x 필터의 너비)
````

````python
class Convolution:
  def __init__(self, W, b, stride =1, pad =0):
    self.W = W
    self.b = b
    self.stride = stride
    self.pad = pad

  def forward(self, x):
    FN, C, FH, FW = self.W.shape
    N, C, H, W = x.shape
    out_h = int(1 + (H + 2*self.pad - FH) / self.stride)
    out_w = int(1 + (W + 2*self.pad - FW) / self.stride)

    col = im2col(x, FH, FW, self.stride, self.pad)
    col_W = self.W.reshape(FN, -1).T # 필터 전개
    out = np.dot(col, col_W) + self.b

    out = out.reshape(N, out_h, out_w, -1).transpose(0, 3, 1, 2)

    return out
````

# 풀링 계층 (Pooling Layer)

풀링(Pooling)은 CNN에서 사용하는 계층으로,

특징 맵(Feature Map)의 크기를 줄이면서 중요한 정보만 남기는 역할을 한다.

보통 합성곱 계층(Convolution Layer) 뒤에 사용한다.

CNN의 일반적인 구조는 다음과 같다.

```text
입력 이미지
↓
Convolution
↓
ReLU
↓
Pooling
↓
Convolution
↓
ReLU
↓
Pooling
↓
Affine
↓
Softmax
```

---

# 1. 풀링 계층이 필요한 이유

합성곱 계층을 여러 번 통과하면 특징 맵의 크기가 매우 커질 수 있다.

예를 들어

$$
224 \times 224
$$

이미지가 있다고 하자.

합성곱을 여러 번 수행하면

```text
특징 맵 개수 증가
↓
연산량 증가
↓
메모리 사용량 증가
```

문제가 발생한다.

---

따라서

```text
중요한 특징만 유지
↓
데이터 크기 감소
```

를 위해 풀링 계층을 사용한다.

---

# 2. 풀링의 기본 아이디어

풀링은 작은 영역(Window)을 선택한 뒤

그 영역을 대표하는 값 하나만 출력한다.

예를 들어

$$
2 \times 2
$$

영역을 하나의 값으로 압축한다.

---

# 13. 풀링의 효과

## 1. 연산량 감소

특징 맵 크기를 줄인다.

---

예시

$$
28\times28
$$

↓

$$

$$

$$

$$

연산량이 크게 감소한다.

---

## 2. 메모리 사용량 감소

특징 맵 크기가 작아지므로

저장해야 하는 데이터도 줄어든다.

---

## 3. 오버피팅 감소

---

## 4. 위치 변화에 강함

예를 들어

고양이 눈이 약간 이동해도


이라고 한다.

CNN은 보통

```text

↓

↓
```

구조를 반복한다.

---

합성곱 계층

```text
특징 추출
```

---

풀링 계층

```text
특징 압축
```

---

즉

```text
Convolution
=
```

# 핵심 정리

1. 풀링 계층은 특징 맵 크기를 줄이는 계층이다.

2. 가장 많이 사용하는 방식은 Max Pooling이다.

3. Max Pooling은 영역 안의 최댓값을 선택한다.

4. Average Pooling은 영역 안의 평균값을 선택한다.

5. 풀링은 높이와 너비만 줄이고 채널 수는 유지한다.

6. 풀링 계층에는 학습해야 할 가중치가 없다.

7. 출력 크기는 다음 식으로 계산한다.

$$
OH
=
\frac{H-PH}{S}+1
$$

$$
OW
=
\frac{W-PW}{S}+1
$$

8. 풀링은 연산량 감소, 메모리 절약, 오버피팅 감소 효과를 가진다.

9. CNN에서 합성곱 계층은 특징 추출, 풀링 계층은 특징 압축 역할을 수행한다.

````python
class Pooling:
  def __init__(self, pool_h, pool_w, stride =1, pad =0):
    self.pool_h = pool_h
    self.pool_w = pool_w
    self.stride = stride
    self.pad = pad

  def forward(self, x):
    N, C, H, W = x.shape
    out_h = int(1 + (H - self.pool_h) / self.stride)
    out_w = int(1 + (W - self.pool_w) / self.stride)

    # 전개 (1)
    col = im2col(x, self.pool_h, self.pool_w, self.stride, self.pad)
    col = col.reshape(-1, self.pool_h*self.pool_w)

    # 최댓값 (2)
    out = np.max(col, axis=1)
    # 성형 (3)
    out = out.reshape(N, out_h, out_w, C).transpose(0,3,1,2)

    return out
````

````python
#CNN 구현
class SimpleConvNet:
  def __init__(self, input_dim=(1, 28, 28),
               conv_param={'filter_num':30, 'filter_size':5, 'pad':0, 'stride':1},
               hidden_size = 100, output_size = 10, weight_init_std = 0.01):
    filter_num = conv_param['filter_num']
    filter_size = conv_param['filter_size']
    filter_pad = conv_param['pad']
    filter_stride = conv_param['stride']
    input_size = input_dim[1]
    conv_output_size = (input_size - filter_size + 2*filter_pad) / filter_stride + 1
    pool_output_size = int(filter_num * (conv_output_size/2) * (conv_output_size/2))

    self.params = {}
    self.params['W1'] = weight_init_std * np.random.randn(filter_num, input_dim[0], filter_size, filter_size)
    self.params['b1'] = np.zeros(filter_num)
    self.params['W2'] = weight_init_std * np.random.rand(pool_output_size, hidden_size)
    self.params['b2'] = np.zeros(hidden_size)
    self.params['W3'] = weight_init_std * np.random.rand(hidden_size, output_size)
    self.params['b3'] = np.zeros(output_size)

    self.layers = OrderedDict()
    self.layers['Conv1'] = Convolution(self.params['W1'], self.params['b1'], conv_param['stride'], conv_param['pad'])
    self.layers['Relu1'] = Relu()
    self.layers['Pool1'] = Pooling(pool_h=2, pool_w=2, stride=2)
    self.layers['Affine1'] = Affine(self.params['W2'], self.params['b2'])
    self.layers['ReLu2'] = Relu()
    self.layers['Affine2'] = Affine(self.params['W3'], self.params['b3'])
    self.last_layer = SoftmaxWithLoss()

  def predict(self, x):
    for layer in self.layers.values():
      x = layer.forward(x)
    return x

  def loss(self, x, t):
    y = self.predict(x)
    return self.last_layer.forward(y, t)

  def gradient(self, x, t):
    self.loss(x, t)

    dout = 1
    dout = self.last_layer.backward(dout)

    layers = list(self.layers.values())
    layers.reverse()
    for layer in layers:
      dout = layer.backward(dout)

    grads ={}
    grads['W1'] = self.layers['Conv1'].dW
    grads['b1'] = self.layers['Conv1'].db
    grads['W2'] = self.layers['Affine1'].dW
    grads['b2'] = self.layers['Affine1'].db
````
