# 최적화 알고리즘 (Optimization Algorithm)

딥러닝의 목표는 손실 함수(Loss Function)를 최소화하는 가중치(Weight)를 찾는 것이다.

이를 위해 사용하는 대표적인 최적화 알고리즘은 다음과 같다.

1. 확률적 경사 하강법 (SGD)
2. 모멘텀 (Momentum)
3. AdaGrad
4. Adam

---

# 1. 확률적 경사 하강법 (Stochastic Gradient Descent, SGD)

## 개념

확률적 경사 하강법(SGD)은 손실 함수의 기울기(Gradient)를 이용하여 가중치를 조금씩 수정하는 방법이다.

손실 함수가 가장 빠르게 감소하는 방향은 기울기의 반대 방향이므로, 가중치를 기울기의 반대 방향으로 이동시킨다.

---

## SGD 업데이트 식

$$
W \leftarrow W - \eta \nabla L
$$

여기서

- $W$ : 가중치(Weight)
- $\eta$ : 학습률(Learning Rate)
- $\nabla L$ : 손실 함수의 기울기(Gradient)

이다.

---

## SGD 동작 과정

```text
현재 가중치
↓
기울기 계산
↓
기울기 반대 방향 이동
↓
가중치 업데이트
↓
반복
```

---

## SGD의 직관적 이해

산 정상에서 가장 낮은 계곡을 찾는다고 생각해보자.

현재 위치에서

```text
어느 방향으로 내려가야
가장 빠르게 내려갈 수 있을까?
```

를 계산한 후

조금 이동한다.

그리고 다시 계산한다.

이를 반복하면서 최저점에 도달한다.

---

# SGD의 단점

## 1. 진동(Oscillation)

손실 함수의 모양이 길쭉한 골짜기 형태인 경우

SGD는 최적점을 향해 직진하지 못한다.

```text
↘
 ↗
↘
 ↗
```

처럼 좌우로 흔들리며 이동한다.

---

## 2. 수렴 속도가 느림

매번 현재 기울기만 보고 움직인다.

이전 방향 정보를 기억하지 못한다.

따라서 최적점까지 도달하는 시간이 길어진다.

---

## 3. 지역 최솟값(Local Minimum)

손실 함수가 복잡한 경우

전역 최솟값(Global Minimum)이 아닌

지역 최솟값(Local Minimum)에 갇힐 수 있다.

---

## 4. 학습률 설정이 어려움

학습률이 너무 크면

```text
최적점을 지나침
```

학습률이 너무 작으면

```text
학습이 매우 느림
```

---

# 2. 모멘텀 (Momentum)

## 등장 배경

SGD는 현재 기울기만 사용한다.

즉,

```text
과거 이동 방향을 기억하지 못함
```

이라는 문제가 있다.

이를 해결하기 위해 모멘텀을 사용한다.

---

## 개념

모멘텀은

```text
이전 이동 방향을 기억
+
현재 기울기 반영
```

하는 방식이다.

물리학의 관성(Inertia) 개념과 비슷하다.

공이 굴러가면 바로 멈추지 않고 계속 굴러가는 것과 같다.

---

## 모멘텀 수식

속도(Velocity)를 정의한다.

### 속도 업데이트

$$
v_t = \beta v_{t-1} - \eta \nabla L
$$

---

### 가중치 업데이트

$$
W_t = W_{t-1} + v_t
$$

---

## 변수 의미

- $v_t$ : 현재 속도
- $v_{t-1}$ : 이전 속도
- $\beta$ : 모멘텀 계수 (보통 0.9)
- $\eta$ : 학습률
- $\nabla L$ : 기울기

---

## 장점

### 1. 진동 감소

좌우로 흔들리는 현상이 줄어든다.

---

### 2. 수렴 속도 증가

이전 방향을 기억하기 때문에 더 빠르게 이동한다.

---

### 3. 지역 최솟값 탈출 가능성 증가

관성 효과로 작은 지역 최솟값을 넘어갈 수 있다.

---

# 3. AdaGrad (Adaptive Gradient)

## 등장 배경

SGD는 모든 파라미터에 동일한 학습률을 사용한다.

하지만 실제로는

```text
자주 업데이트되는 가중치
```

와

```text
거의 업데이트되지 않는 가중치
```

가 존재한다.

---

## 개념

각 가중치마다 다른 학습률을 사용한다.

많이 학습된 가중치는

```text
학습률 감소
```

적게 학습된 가중치는

```text
학습률 유지
```

한다.

---

## 기울기 누적

$$
h_t = h_{t-1} + (\nabla L)^2
$$

---

## 가중치 업데이트

$$
W_t
=
W_{t-1}
-
\frac{\eta}{\sqrt{h_t+\epsilon}}
\nabla L
$$

---

## 변수 의미

- $h_t$ : 기울기 누적값
- $\epsilon$ : 0으로 나누는 것을 방지하는 작은 값

---

## AdaGrad의 특징

기울기가 자주 발생할수록

$$
h_t
$$

가 커진다.

그러면

$$
\frac{\eta}{\sqrt{h_t}}
$$

가 작아진다.

즉,

```text
학습률이 자동 감소
```

한다.

---

## 장점

### 학습률 자동 조절

가중치마다 적절한 학습률 사용 가능

---

## 단점

기울기를 계속 누적하기 때문에

$$
h_t
$$

가 계속 증가한다.

결국

$$
\frac{\eta}{\sqrt{h_t}}
\rightarrow 0
$$

가 된다.

즉,

```text
학습이 거의 멈춤
```

현상이 발생한다.

---

# 4. Adam (Adaptive Moment Estimation)

## 등장 배경

Momentum과 AdaGrad의 장점을 합친 알고리즘이다.

현재 가장 많이 사용되는 최적화 알고리즘 중 하나이다.

---

## 핵심 아이디어

Adam은

```text
Momentum
+
AdaGrad
```

를 결합한 방식이다.

---

## 1차 모멘트

기울기의 평균을 계산한다.

$$
m_t
=
\beta_1 m_{t-1}
+
(1-\beta_1)\nabla L
$$

---

## 2차 모멘트

기울기 제곱의 평균을 계산한다.

$$
v_t
=
\beta_2 v_{t-1}
+
(1-\beta_2)(\nabla L)^2
$$

---

## 편향 보정

초기에는 값이 작아지는 문제가 있기 때문에 보정한다.

### 1차 모멘트 보정

$$
\hat m_t
=
\frac{m_t}
{1-\beta_1^t}
$$

---

### 2차 모멘트 보정

$$
\hat v_t
=
\frac{v_t}
{1-\beta_2^t}
$$

---

## 최종 업데이트 식

$$
W_t
=
W_{t-1}
-
\eta
\frac{\hat m_t}
{\sqrt{\hat v_t}+\epsilon}
$$

---

## 변수 의미

- $m_t$ : 기울기 평균
- $v_t$ : 기울기 제곱 평균
- $\beta_1$ : 1차 모멘트 계수 (보통 0.9)
- $\beta_2$ : 2차 모멘트 계수 (보통 0.999)
- $\epsilon$ : 작은 값

---

## Adam의 장점

### 1. 빠른 수렴

Momentum 효과로 빠르게 이동

---

### 2. 학습률 자동 조절

AdaGrad 효과 활용

---

### 3. 안정적 학습

진동이 적음

---

### 4. 하이퍼파라미터에 덜 민감

초보자가 사용하기 쉽다.

---

# 최적화 알고리즘 비교

| 알고리즘 | 특징 | 장점 | 단점 |
|-----------|--------|--------|--------|
| SGD | 기본 경사 하강법 | 구현 간단 | 느리고 진동 발생 |
| Momentum | 관성 추가 | 빠른 수렴 | 학습률 설정 필요 |
| AdaGrad | 학습률 자동 조절 | 희소 데이터에 강함 | 학습률이 너무 작아짐 |
| Adam | Momentum + AdaGrad | 가장 많이 사용 | 메모리 사용량 증가 |

---

# 핵심 정리

1. SGD는 기울기 방향으로 가중치를 업데이트한다.

$$
W \leftarrow W-\eta\nabla L
$$

2. SGD는 진동과 느린 수렴 문제가 있다.

3. Momentum은 이전 이동 방향을 기억하여 수렴 속도를 높인다.

$$
v_t=\beta v_{t-1}-\eta\nabla L
$$

4. AdaGrad는 가중치마다 다른 학습률을 사용한다.

$$
W_t
=
W_{t-1}
-
\frac{\eta}{\sqrt{h_t+\epsilon}}
\nabla L
$$

5. Adam은 Momentum과 AdaGrad를 결합한 방식이다.

$$
W_t
=
W_{t-1}
-
\eta
\frac{\hat m_t}
{\sqrt{\hat v_t}+\epsilon}
$$

6. 현재 딥러닝에서는 Adam이 가장 널리 사용되는 최적화 알고리즘 중 하나이다.

````python
import numpy as np

class SGD:
  def __init__(self, lr = 0.01):
    self.lr = lr

  def update(self, params, grads):
    for key in params.keys():
      params[key] -= self.lr * grads[key]

class Momentum:
  def __init__(self, lr = 0.01, momentum = 0.9):
    self.lr = lr
    self.momentum = momentum
    self.v = None

  def update(self, params, grads):
    if self.v is None:
      self.v = {}
      for key, val in params.item():
        self.v[key] = np.zeros_like(val)

    for key in params.keys():
      self.v[key] = self.momentum * self.v[key] - self.lr * grads[key]
      params[key] += self.v[key]

class AdaGrad:
  def __init__(self, lr = 0.01):
    self.lr = lr
    self.h = None

  def update(self, params, grads):
    if self.h is None:
      self.h = {}
      for key, val in params.items():
        self.h[key] = np.zeros_like(val)

      for key in params.keys():
        self.h[key] += grads[key] * grads[key]
        params[key] -= self.lr * grads[key] / (np.sqrt(self.h[key]) + 1e-7)
````

# 가중치 초기값 (Weight Initialization)

신경망 학습에서 가중치의 초기값은 매우 중요하다.

초기값을 잘못 설정하면 학습이 매우 느려지거나 학습 자체가 제대로 이루어지지 않을 수 있다.

---

# 가중치 초기값을 0으로 하면 안 되는 이유

가중치를 모두 0으로 초기화한다고 가정하자.

$$
W =
\begin{bmatrix}
0 & 0\\
0 & 0
\end{bmatrix}
$$

---

## 문제점 : 대칭성(Symmetry)

은닉층의 모든 뉴런이 동일한 값을 출력한다.

예를 들어

$$
z_1 = w_1x+b
$$

$$
z_2 = w_2x+b
$$

인데

$$
w_1=w_2=0
$$

이면

$$
z_1=z_2
$$

가 된다.

---

결과적으로

```text
모든 뉴런이 같은 출력
↓
모든 뉴런이 같은 기울기
↓
같은 방식으로 업데이트
↓
뉴런들이 서로 다른 특징을 학습하지 못함
```

---

## 결론

가중치는 반드시 무작위(Random)로 초기화해야 한다.

예시

```python
W = np.random.randn(2, 2)
```

---

# 은닉층의 활성화값 분포

깊은 신경망에서는 각 층의 출력이 다음 층의 입력이 된다.

예를 들어

```text
입력층
↓
은닉층1
↓
은닉층2
↓
은닉층3
↓
출력층
```

---

활성화값이 너무 크면

$$
z \gg 0
$$

가 되고,

너무 작으면

$$
z \ll 0
$$

가 된다.

---

이 경우

Sigmoid 함수가 포화 영역(Saturation Region)에 들어간다.

$$
\sigma(z)=\frac{1}{1+e^{-z}}
$$

---

포화 영역에서는

$$
\sigma'(z)\approx0
$$

이 된다.

즉

```text
기울기 소실(Vanishing Gradient)
```

이 발생한다.

---

## 좋은 활성화값 분포

좋은 신경망은

```text
너무 크지도 않고
너무 작지도 않은
```

분포를 가져야 한다.

보통 평균이 0 근처이고 적절한 분산을 가지도록 초기화한다.

---

# Xavier 초기값 (Xavier Initialization)

## 등장 배경

Sigmoid와 Tanh는 입력값이 커질수록 포화된다.

따라서 층을 지날수록 분산이 커지거나 작아지지 않도록 초기화해야 한다.

---

## Xavier 초기화

가중치를 다음과 같이 초기화한다.

$$
W \sim N
\left(
0,
\frac{1}{n}
\right)
$$

또는

$$
W =
\frac{1}{\sqrt{n}}
\times
N(0,1)
$$

---

여기서

- $n$ : 이전 층의 노드 수

이다.

---

## 예시

입력 노드가 100개라면

$$
\sqrt{100}=10
$$

이므로

```python
W = np.random.randn(100, 50) / 10
```

으로 초기화한다.

---

## Xavier 초기화가 적합한 경우

- Sigmoid
- Tanh

---

# ReLU를 사용할 때 가중치 초기값

## 문제점

ReLU는 Sigmoid와 다른 특성을 가진다.

$$
f(x)=\max(0,x)
$$

음수는 모두 0이 된다.

---

따라서 Xavier 초기화보다 더 큰 분산이 필요하다.

---

# He 초기값 (He Initialization)

ReLU 전용 초기화 방법이다.

$$
W \sim N
\left(
0,
\frac{2}{n}
\right)
$$

또는

$$
W =
\sqrt{\frac{2}{n}}
\times N(0,1)
$$

---

## 예시

입력 노드 수가 100개라면

$$
\sqrt{\frac{2}{100}}
=
0.141
$$

따라서

```python
W = np.random.randn(100,50) * 0.141
```

로 초기화한다.

---

## He 초기화가 적합한 경우

- ReLU
- Leaky ReLU
- ELU

---

# 배치 정규화 (Batch Normalization)

## 등장 배경

신경망이 깊어질수록 각 층의 입력 분포가 계속 변한다.

이를

```text
Internal Covariate Shift
```

라고 한다.

---

이 문제를 해결하기 위해 Batch Normalization을 사용한다.

---

# 배치 정규화 알고리즘

미니배치 데이터

$$
x_1,x_2,\cdots,x_m
$$

에 대해

---

## 1단계 : 평균 계산

$$
\mu_B
=
\frac{1}{m}
\sum_{i=1}^{m}x_i
$$

---

## 2단계 : 분산 계산

$$
\sigma_B^2
=
\frac{1}{m}
\sum_{i=1}^{m}
(x_i-\mu_B)^2
$$

---

## 3단계 : 정규화

$$
\hat{x}_i
=
\frac{x_i-\mu_B}
{\sqrt{\sigma_B^2+\epsilon}}
$$

---

## 4단계 : 스케일링과 이동

$$
y_i
=
\gamma \hat{x}_i+\beta
$$

---

여기서

- $\gamma$ : 스케일 파라미터
- $\beta$ : 이동 파라미터

이다.

---

# 배치 정규화의 효과

## 1. 학습 속도 향상

더 큰 학습률 사용 가능

---

## 2. 기울기 소실 감소

활성화값이 적절한 범위를 유지한다.

---

## 3. 초기값 의존성 감소

초기값에 덜 민감하다.

---

## 4. 일반화 성능 향상

약한 정규화 효과를 가진다.

---

# 오버피팅 (Overfitting)

## 정의

훈련 데이터에만 지나치게 맞춘 상태

---

## 특징

```text
훈련 정확도 ↑
테스트 정확도 ↓
```

---

## 원인

- 데이터 부족
- 모델이 너무 복잡
- Epoch 과다

# 가중치 감소 (Weight Decay)

## 개념

가중치가 너무 커지는 것을 막는 방법

---

## 목적

복잡한 모델을 단순하게 만들어 과적합을 줄인다.

---

## 손실 함수

원래 손실 함수

$$
L
$$

에

가중치 크기 패널티를 추가한다.

$$
L'
=
L
+
\frac{\lambda}{2}
\sum W^2
$$

# 핵심 정리

1. 가중치를 0으로 초기화하면 모든 뉴런이 동일하게 학습된다.

2. Sigmoid와 Tanh는 Xavier 초기화를 사용한다.

$$
W=
\frac{1}{\sqrt{n}}
N(0,1)
$$

3. ReLU는 He 초기화를 사용한다.

$$
W=
\sqrt{\frac{2}{n}}
N(0,1)
$$

4. 배치 정규화는 입력 분포를 정규화한다.

5. 가중치 감소는 큰 가중치를 억제한다.

$$
L'
=
L+\frac{\lambda}{2}\sum W^2
$$

6. 드롭아웃은 일부 뉴런을 무작위로 제거한다.

7. 검증 데이터는 하이퍼파라미터 선택에 사용한다.

8. Random Search는 효율적인 하이퍼파라미터 탐색 방법이다.
