# 퍼셉트론 (Perceptron)

퍼셉트론은 1957년 프랭크 로젠블랫(Frank Rosenblatt)이 제안한 가장 기본적인 인공신경망 모델이다. 여러 입력값을 받아 하나의 출력을 생성하며, 현재 딥러닝의 기초가 되는 개념이다.

---

## 퍼셉트론의 구조

퍼셉트론은 다음 요소로 구성된다.

- 입력값(Input): $x_1, x_2, \cdots, x_n$
- 가중치(Weight): $w_1, w_2, \cdots, w_n$
- 편향(Bias): $b$
- 활성화 함수(Activation Function)

각 입력값은 가중치와 곱해진 후 모두 더해지고, 편향이 추가된다.

$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

행렬 형태로 표현하면

$$
z = \mathbf{w}^T\mathbf{x} + b
$$

이다.

---

## 가중치 (Weight)

가중치는 각 입력값의 중요도를 나타낸다.

$$
w_i x_i
$$

- 가중치가 클수록 해당 입력의 영향력이 커진다.
- 가중치가 작을수록 해당 입력의 영향력이 작아진다.
- 학습 과정에서 가중치가 계속 수정된다.

예를 들어

$$
z = 0.8x_1 + 0.2x_2
$$

라면 $x_1$이 $x_2$보다 4배 더 중요한 특성으로 사용된다고 볼 수 있다.

---

## 편향 (Bias)

편향은 뉴런이 얼마나 쉽게 활성화될지를 결정하는 값이다.

$$
z = \sum_{i=1}^{n} w_i x_i + b
$$

편향은 결정 경계(Decision Boundary)를 이동시키는 역할을 한다.

예를 들어

$$
z = x_1 + x_2
$$

와

$$
z = x_1 + x_2 - 5
$$

는 동일한 기울기를 가지지만 분류 기준이 달라진다.

따라서 편향은 모델이 더욱 유연한 분류를 수행할 수 있도록 도와준다.

---

## 활성화 함수

계산된 값 $z$는 활성화 함수를 통과하여 최종 출력값을 생성한다.

초기 퍼셉트론에서는 계단 함수(Step Function)를 사용하였다.

$$
y =
\begin{cases}
1, & z > 0 \\\\
0, & z \le 0
\end{cases}
$$

즉, 계산 결과가 0보다 크면 1을 출력하고, 그렇지 않으면 0을 출력한다.

---

## 퍼셉트론의 동작 과정

1. 입력값을 받는다.
2. 입력값과 가중치를 곱한다.
3. 모든 결과를 더한다.
4. 편향을 더한다.
5. 활성화 함수를 적용한다.
6. 최종 출력을 생성한다.

수식으로 표현하면

$$
y = f(\mathbf{w}^T\mathbf{x} + b)
$$

여기서

- $\mathbf{x}$ : 입력 벡터
- $\mathbf{w}$ : 가중치 벡터
- $b$ : 편향
- $f$ : 활성화 함수

이다.

---

## 퍼셉트론의 한계

단층 퍼셉트론(Single Layer Perceptron)은 선형적으로 분리 가능한 문제만 해결할 수 있다.

예를 들어

- AND 게이트 : 가능
- OR 게이트 : 가능
- XOR 게이트 : 불가능

XOR 문제를 해결하기 위해 은닉층(Hidden Layer)을 추가한 다층 퍼셉트론(Multi-Layer Perceptron, MLP)이 등장하였고, 이는 현대 딥러닝의 기반이 되었다.

# 활성화 함수 (Activation Function)

## 활성화 함수란?

활성화 함수(Activation Function)는 뉴런이 입력값을 받아 계산한 결과를 다음 층으로 전달할지 결정하는 함수이다.

퍼셉트론에서 입력값과 가중치, 편향을 이용하여 다음과 같은 값을 계산한다.

$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

이 계산 결과 $z$를 그대로 사용하는 것이 아니라 활성화 함수 $f$에 통과시켜 최종 출력값을 생성한다.

$$
y = f(z)
$$

즉, 활성화 함수는 뉴런의 출력을 결정하는 역할을 수행한다.

---

## 활성화 함수를 사용하는 이유

신경망은 여러 층(Layer)을 쌓아 복잡한 패턴을 학습한다.

만약 활성화 함수로 선형 함수만 사용한다면,

$$
y = ax + b
$$

형태의 계산만 반복하게 된다.

예를 들어,

$$
y_1 = a_1x + b_1
$$

$$
y_2 = a_2y_1 + b_2
$$

를 계산하면

$$
y_2 = a_2(a_1x+b_1)+b_2
$$

$$
y_2 = (a_1a_2)x + (a_2b_1+b_2)
$$

가 된다.

즉, 여러 층을 쌓아도 결국 하나의 선형 함수와 동일해진다.

따라서 복잡한 곡선이나 비선형 패턴을 학습할 수 없다.

---

## 활성화 함수에서 비선형 함수를 사용해야 하는 이유

딥러닝의 핵심은 복잡한 비선형 관계를 학습하는 것이다.

비선형 활성화 함수를 사용하면 신경망은 다음과 같은 관계를 학습할 수 있다.

- 이미지 분류
- 음성 인식
- 자연어 처리
- 객체 탐지

즉,

**"층을 깊게 쌓는 의미를 만들기 위해 반드시 비선형 함수를 사용해야 한다."**

---

# 1. 계단 함수 (Step Function)

퍼셉트론에서 사용된 가장 초기의 활성화 함수이다.

수식은 다음과 같다.

$$
f(x)=
\begin{cases}
1, & x > 0 \\
0, & x \le 0
\end{cases}
$$

### 특징

- 입력이 0보다 크면 1 출력
- 입력이 0 이하이면 0 출력
- 출력값이 이진값(0 또는 1)
- 퍼셉트론에서 사용
- 미분이 불가능하여 딥러닝 학습에는 사용되지 않음

---

# 2. 시그모이드 함수 (Sigmoid Function)

계단 함수를 부드럽게 만든 함수이다.

수식은 다음과 같다.

$$
f(x)=\frac{1}{1+e^{-x}}
$$

### 특징

- 출력 범위가 0 ~ 1
- 출력값을 확률처럼 해석 가능
- 함수가 매끄럽고 미분 가능
- 과거 신경망에서 널리 사용

### 장점

- 미분 가능
- 확률 해석 가능

### 단점

- 입력값이 매우 크거나 작으면 기울기가 거의 0이 됨
- 기울기 소실(Vanishing Gradient) 문제 발생
- 학습 속도가 느림

---

# 3. ReLU 함수 (Rectified Linear Unit)

현재 딥러닝에서 가장 많이 사용되는 활성화 함수이다.

수식은 다음과 같다.

$$
f(x)=\max(0,x)
$$

또는

$$
f(x)=
\begin{cases}
x, & x > 0 \\
0, & x \le 0
\end{cases}
$$

### 특징

- 양수는 그대로 출력
- 음수는 0 출력
- 계산이 매우 단순
- 학습 속도가 빠름

### 장점

- 계산량이 적음
- 기울기 소실 문제를 완화
- 딥러닝에서 높은 성능

### 단점

- 음수 영역에서는 항상 0 출력
- 일부 뉴런이 죽는 Dying ReLU 문제 발생 가능

---

# 활성화 함수 비교

| 함수 | 수식 | 출력 범위 | 장점 | 단점 |
|--------|--------|--------|--------|--------|
| 계단 함수 | $0$ 또는 $1$ | 0, 1 | 구조가 단순 | 미분 불가 |
| 시그모이드 | $\frac{1}{1+e^{-x}}$ | 0 ~ 1 | 확률 해석 가능 | 기울기 소실 |
| ReLU | $\max(0,x)$ | 0 ~ $\infty$ | 빠른 학습, 높은 성능 | Dying ReLU |

---

## 정리

활성화 함수는 뉴런의 출력값을 결정하는 함수이다.

$$
y = f(\mathbf{w}^T\mathbf{x}+b)
$$

신경망이 복잡한 패턴을 학습하기 위해서는 비선형 활성화 함수가 반드시 필요하다.

현재 딥러닝에서는 대부분 ReLU 계열 함수가 사용되며, 시그모이드 함수는 출력층에서 확률을 표현할 때 주로 사용된다.

```python
import numpy as np
import matplotlib.pylab as plt


#계단함수
def step_function(x):
	y = x > 0
	return y.astype(int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()


```

```python
#Sigmoid 함수
def sigmoid(x):
	return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
#ReLU 함수
def relu(x):
	return np.maximum(0,x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```

# 행렬의 곱 (Matrix Multiplication)

신경망에서는 행렬의 곱셈이 매우 중요하다.

NumPy에서는 np.dot() 함수를 사용하여 행렬 곱을 수행한다.

python np.dot(A, B)

---

## 행렬 곱의 조건

행렬

$$
A(m,n)
$$

와

$$
B(n,p)
$$

가 있을 때,

앞 행렬의 열 개수와 뒤 행렬의 행 개수가 같아야 한다.

즉,

$$
(m,n)\times(n,p)
$$

형태일 때만 곱셈이 가능하다.

예를 들어

$$
(2,3)\times(3,4)
$$

는 가능하지만

$$
(2,3)\times(2,4)
$$

는 불가능하다.

---

## 행렬 곱 예제

행렬

$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$

와

$$
B =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
$$

를 곱해보자.

행렬 형태로 표현하면

$$
AB =
\begin{bmatrix}
19 & 22 \\
43 & 50
\end{bmatrix}
$$

이다.

---

## 행렬 곱 계산 과정

첫 번째 원소

$$
1\times5 + 2\times7 = 19
$$

두 번째 원소

$$
1\times6 + 2\times8 = 22
$$

세 번째 원소

$$
3\times5 + 4\times7 = 43
$$

네 번째 원소

$$
3\times6 + 4\times8 = 50
$$

이다.

---

## 결과 행렬의 크기

행렬 곱

$$
(m,n)\times(n,p)
$$

의 결과는

$$
(m,p)
$$

크기의 행렬이 된다.

예를 들어

$$
(2,3)\times(3,4)
$$

의 결과는

$$
(2,4)
$$

이다.

$$
X =
\begin{bmatrix}
x_1 & x_2
\end{bmatrix}
$$

$$

\begin{bmatrix}

$$

$$

를 계산한다.

행렬의 크기는

$$
(1,2)\times(2,3)
$$

이므로 결과는

$$
(1,3)
$$

이 된다.

따라서 출력은

$$
Y =
\begin{bmatrix}
y_1 & y_2 & y_3
\end{bmatrix}
$$

가 된다.

#신경망 구현

```python
def init_network():
    network = {}
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([0.1, 0.2])

    return network


def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = identity_function(a3)

    return y


network = init_network()
x = np.array([1.0, 0.5])
y = forward(network, x)  # [ 0.31682708  0.69627909]

print(y)
```

# Softmax 함수

## Softmax 함수란?

Softmax 함수는 신경망의 출력값을 확률(Probability) 형태로 변환하는 함수이다.

주로 다중 분류(Multi-class Classification) 문제의 출력층(Output Layer)에서 사용된다.

신경망의 출력값이

$$
a_1, a_2, \cdots, a_n
$$

일 때 Softmax 함수는 각 클래스에 속할 확률을 계산한다.

---

## Softmax 함수의 수식

$$
y_k=
\frac{e^{a_k}}
{\sum_{i=1}^{n}e^{a_i}}
$$

여기서

- $a_k$ : k번째 입력값(점수)
- $y_k$ : k번째 클래스의 확률
- $e$ : 자연상수

이다.

---

## Softmax 함수의 특징

### 1. 출력값이 모두 0과 1 사이이다.

지수 함수는 항상 양수이므로

$$
0 < y_k < 1
$$

을 만족한다.

따라서 출력값을 확률처럼 해석할 수 있다.

---

### 2. 출력값의 총합이 1이다.

Softmax 함수의 가장 중요한 특징은

$$
\sum_{k=1}^{n} y_k = 1
$$

이라는 점이다.

따라서 각 출력값은 전체 클래스 중 해당 클래스에 속할 확률을 의미한다.

---

### 3. 가장 큰 입력값이 가장 큰 확률을 가진다.

입력값 중 가장 큰 값은 Softmax를 거친 후에도 가장 큰 확률을 가진다.

따라서 최종 예측 클래스는 일반적으로 가장 높은 확률을 가진 클래스로 선택한다.

---

### 4. 다중 분류 문제에서 사용된다.

예를 들어 숫자 인식 문제에서

- 0
- 1
- 2
- ...
- 9

중 하나를 선택해야 할 때 Softmax 함수를 사용한다.

---

### 5. 출력층에서 사용된다.

은닉층(Hidden Layer)에서는 주로 ReLU 함수를 사용하고,

출력층(Output Layer)에서는 Softmax 함수를 사용하여 최종 결과를 확률로 변환한다.

---

### 6. 오버플로우 문제를 방지해야 한다.

입력값이 매우 크면

$$
e^{1000}
$$

과 같은 값이 발생하여 오버플로우가 발생할 수 있다.

이를 방지하기 위해 입력값 중 최댓값을 빼서 계산한다.

$$
y_k=
\frac{e^{a_k-C}}
{\sum_{i=1}^{n}e^{a_i-C}}
$$

여기서

$$
C=\max(a)
$$

이다.

---

## 정리

- Softmax 함수는 출력값을 확률로 변환하는 함수이다.
- 다중 분류 문제의 출력층에서 사용된다.
- 출력값은 모두 0과 1 사이의 값을 가진다.
- 출력값의 총합은 항상 1이다.
- 가장 큰 입력값은 가장 큰 확률로 변환된다.
- 오버플로우를 방지하기 위해 최댓값을 빼고 계산한다.

$$
y_k=
\frac{e^{a_k}}
{\sum_{i=1}^{n}e^{a_i}}
$$

```python
#softmax 함수
def softmax(a):
  c = np.max(a)
  exp_a = np.exp(a -c)
  sum_exp_a = np.sum(exp_a)
  y = exp_a / sum_exp_a

  return y
```

```python
a = np.array([0.3, 2.9, 4.0])
y = softmax(a)
print(y)
print(np.sum(y))
```

```python
import sys
import pickle
import numpy as np

# 책 예제 코드 폴더 경로 추가
sys.path.append('/content/deep-learning-from-scratch')

from dataset.mnist import load_mnist
from common.functions import sigmoid, softmax


# MNIST 데이터 불러오기
def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(
        flatten=True,
        normalize=True,
        one_hot_label=False
    )
    return x_test, t_test


# 학습된 가중치 불러오기
def init_network():
    with open('/content/deep-learning-from-scratch/ch03/sample_weight.pkl', 'rb') as f:
        network = pickle.load(f)
    return network


# 신경망 추론 함수
def predict(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)

    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)

    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)

    return y


# 데이터와 신경망 준비
x, t = get_data()
network = init_network()

accuracy_cnt = 0
batch_size = 100

# 배치 처리
for i in range(0, len(x), batch_size):
    x_batch = x[i:i + batch_size]
    y_batch = predict(network, x_batch)

    p = np.argmax(y_batch, axis=1)
    accuracy_cnt += np.sum(p == t[i:i + batch_size])

# 정확도 출력
print("Accuracy:" + str(float(accuracy_cnt) / len(x)))
```
